```python
#Backword
```

```python
def backward(X, y_true, z1, a1, z2, a2):
    n = X.shape[0]

    # Output layer gradient: softmax + cross-entropy combined derivative
    # This simplifies neatly to just (a2 - y_true) — a famous result
    dz2 = (a2 - y_true) / n                # shape: (n_samples, n_output)

    # Gradients for W2 and b2
    dW2 = a1.T @ dz2                        # shape: (n_hidden, n_output)
    db2 = np.sum(dz2, axis=0, keepdims=True)

    # Backprop through ReLU: multiply by derivative of ReLU
    da1 = dz2 @ W2.T                        # shape: (n_samples, n_hidden)
    dz1 = da1 * relu_derivative(a1)        # shape: same

    # Gradients for W1 and b1
    dW1 = X.T @ dz1                         # shape: (n_input, n_hidden)
    db1 = np.sum(dz1, axis=0, keepdims=True)

    return dW1, db1, dW2, db2
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder

# ─── 1. LOAD & PREPARE DATA ───────────────────────────────────────────────────

iris = load_iris()
X = iris.data        # shape (150, 4) — 150 samples, 4 features
y = iris.target      # shape (150,)  — values 0, 1, 2

# Normalize: subtract mean, divide by std — keeps values small for stable training
X_mean = np.mean(X, axis=0)
X_std  = np.std(X, axis=0)
X_norm = (X - X_mean) / X_std

# One-hot encode labels: [0,1,2] → [[1,0,0],[0,1,0],[0,0,1]]
encoder = OneHotEncoder(sparse_output=False)
y_ohe = encoder.fit_transform(y.reshape(-1, 1))   # shape (150, 3)

# Train/test split — 80% train, 20% test
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y_ohe, test_size=0.2, random_state=42
)

print("X_train shape:", X_train.shape)   # (120, 4)
print("y_train shape:", y_train.shape)   # (120, 3)

# ─── 2. NETWORK ARCHITECTURE ──────────────────────────────────────────────────

n_input  = 4    # number of input features
n_hidden = 16   # hidden layer neurons — tune this if needed
n_output = 3    # number of classes (Iris has 3)

# ─── 3. WEIGHT INITIALIZATION ─────────────────────────────────────────────────
# Small random values so activations don't saturate immediately

np.random.seed(42)
W1 = np.random.randn(n_input, n_hidden) * 0.01   # (4, 16)
b1 = np.zeros((1, n_hidden))                      # (1, 16)
W2 = np.random.randn(n_hidden, n_output) * 0.01  # (16, 3)
b2 = np.zeros((1, n_output))                      # (1, 3)

# ─── 4. ACTIVATION FUNCTIONS ──────────────────────────────────────────────────

def relu(z):
    # Zero out all negative values, keep positives unchanged
    return np.maximum(0, z)

def relu_derivative(z):
    # Gradient of ReLU: 1 where input was positive, 0 elsewhere
    return (z > 0).astype(float)

def softmax(z):
    # Subtract max per row for numerical stability before exponentiating
    shifted = z - np.max(z, axis=1, keepdims=True)
    exp_z   = np.exp(shifted)
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

# ─── 5. LOSS FUNCTION ─────────────────────────────────────────────────────────

def cross_entropy_loss(y_true, y_pred):
    # Clip to avoid log(0) blowing up to -inf
    y_pred = np.clip(y_pred, 1e-12, 1 - 1e-12)
    n = y_true.shape[0]
    # Average negative log likelihood over all samples
    return -np.sum(y_true * np.log(y_pred)) / n

# ─── 6. FORWARD PASS ──────────────────────────────────────────────────────────

def forward(X):
    # Layer 1: linear → ReLU
    z1 = X @ W1 + b1       # (n_samples, 16)
    a1 = relu(z1)           # (n_samples, 16)

    # Layer 2: linear → softmax
    z2 = a1 @ W2 + b2      # (n_samples, 3)
    a2 = softmax(z2)        # (n_samples, 3) — probabilities summing to 1

    return z1, a1, z2, a2

# ─── 7. BACKWARD PASS ─────────────────────────────────────────────────────────

def backward(X, y_true, z1, a1, z2, a2):
    n = X.shape[0]

    # Output layer: softmax + cross-entropy gradient simplifies to (a2 - y_true)
    dz2 = (a2 - y_true) / n          # (n_samples, 3)
    dW2 = a1.T @ dz2                  # (16, 3)
    db2 = np.sum(dz2, axis=0, keepdims=True)  # (1, 3)

    # Hidden layer: backprop through ReLU
    da1 = dz2 @ W2.T                  # (n_samples, 16)
    dz1 = da1 * relu_derivative(z1)   # (n_samples, 16)
    dW1 = X.T @ dz1                   # (4, 16)
    db1 = np.sum(dz1, axis=0, keepdims=True)  # (1, 16)

    return dW1, db1, dW2, db2

# ─── 8. TRAINING LOOP ─────────────────────────────────────────────────────────

learning_rate = 0.1
epochs        = 1000
loss_history  = []

for epoch in range(epochs):
    # Forward pass
    z1, a1, z2, a2 = forward(X_train)

    # Compute loss
    loss = cross_entropy_loss(y_train, a2)
    loss_history.append(loss)

    # Backward pass
    dW1, db1, dW2, db2 = backward(X_train, y_train, z1, a1, z2, a2)

    # Gradient descent weight update
    W1 -= learning_rate * dW1
    b1 -= learning_rate * db1
    W2 -= learning_rate * dW2
    b2 -= learning_rate * db2

    if epoch % 100 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss:.4f}")

# ─── 9. EVALUATE ACCURACY ─────────────────────────────────────────────────────

def predict(X):
    _, _, _, a2 = forward(X)
    # Class with highest probability is the prediction
    return np.argmax(a2, axis=1)

def accuracy(X, y_ohe):
    preds       = predict(X)
    true_labels = np.argmax(y_ohe, axis=1)
    return np.mean(preds == true_labels)

train_acc = accuracy(X_train, y_train)
test_acc  = accuracy(X_test,  y_test)

print(f"\nTrain accuracy: {train_acc * 100:.1f}%")
print(f"Test  accuracy: {test_acc  * 100:.1f}%")

# ─── 10. PLOT LOSS CURVE ──────────────────────────────────────────────────────

plt.figure(figsize=(8, 4))
plt.plot(loss_history, color='purple')
plt.title("Training Loss over Epochs")
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.grid(True)
plt.tight_layout()
plt.show()

# ─── 11. SANITY CHECKS ────────────────────────────────────────────────────────

# 1. Probabilities of one sample must sum to 1
_, _, _, a2_sample = forward(X_train[0:1])
print("\nSample probabilities:", a2_sample)
print("Sum (must be 1.0):", a2_sample.sum())

# 2. Loss at epoch 0 should be near 1.0986 (random 3-class = -log(1/3))
print("\nInitial random loss expected ~1.0986")
print("Check your epoch 0 loss above ↑")

# 3. No negative values after ReLU
z1_check, a1_check, _, _ = forward(X_train[0:1])
print("\nMin value after ReLU (must be >= 0):", a1_check.min())
```

```python
import torch
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import torch.optim as optim
from sklearn.model_selection import train_test_split
```

```python
iris=load_iris()
X=iris.data
Y=iris.target
```

```python
#normalization

Xmean=np.mean(X,axis=0)
Xstd=np.std(X,axis=0)
x_norm=(X-Xmean)/Xstd
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x_norm,Y,test_size=0.2,random_state=42
)
```

```python
x_train=torch.tensor(x_train,dtype=torch.float32)
x_test=torch.tensor(x_test,dtype=torch.float32)
y_train=torch.tensor(y_train,dtype=torch.long)
y_test=torch.tensor(y_test,dtype=torch.long)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
#iris Class 
class Iris(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1=nn.Linear(4,16)
        self.layer2=nn.Linear(16,3)
        self.relu=nn.ReLU()
    
    def forward(self,x):
        z1=self.layer1(x)
        a1=self.relu(z1)
        z2=self.layer2(a1)
        return z2
    
```

```python
model=Iris()
```

```python
print("\nModel architecture:")
print(model)
```

```python
# Count parameters — NumPy tracked shapes manually
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params}")
```

```python
criterion=nn.CrossEntropyLoss()
```

```python
optimizer=optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=1000
loss_history=[]
for epoch in range(epochs):
    model.train()
    y_pred=model(x_train)
    
    loss=criterion(y_pred,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    loss_history.append(loss.item())
    
    if epoch%100==0:
        print(f"epoch{epoch:4d}| loss:{loss.item():.4f}")
```

```python
model.eval()
with torch.no_grad():
    train_preds = torch.argmax(model(x_train), dim=1)
    test_preds  = torch.argmax(model(x_test),  dim=1)

train_acc = (train_preds == y_train).float().mean().item()
test_acc  = (test_preds  == y_test).float().mean().item()

print(f"\nTrain accuracy: {train_acc * 100:.1f}%")
print(f"Test  accuracy: {test_acc  * 100:.1f}%")
```

```python
plt.figure(figsize=(8, 4))
plt.plot(loss_history, color='orange', label='PyTorch')
plt.title("Training Loss over Epochs — PyTorch")
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# ─── 1. LOAD & PREPARE DATA ───────────────────────────────────────────────────

iris = load_iris()
X = iris.data
y = iris.target

# Normalize exactly like NumPy version
X_mean = np.mean(X, axis=0)
X_std  = np.std(X, axis=0)
X_norm = (X - X_mean) / X_std

# Train/test split — same random_state so results are comparable
X_train, X_test, y_train, y_test = train_test_split(
    X_norm, y, test_size=0.2, random_state=42
)

# Convert to PyTorch tensors
# NOTE: y stays as integer labels (not one-hot) — PyTorch CrossEntropyLoss handles that internally
X_train = torch.tensor(X_train, dtype=torch.float32)
X_test  = torch.tensor(X_test,  dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.long)
y_test  = torch.tensor(y_test,  dtype=torch.long)

print("X_train shape:", X_train.shape)   # (120, 4)
print("y_train shape:", y_train.shape)   # (120,) — just integers, not one-hot

# ─── 2. DEFINE THE MODEL ──────────────────────────────────────────────────────
# Same architecture: 4 → 16 → 3
# In NumPy we wrote W1, b1, W2, b2 manually
# PyTorch does all of that inside nn.Linear automatically

class IrisNet(nn.Module):
    def __init__(self):
        super().__init__()
        # nn.Linear(in, out) = your W and b combined in one object
        self.layer1 = nn.Linear(4, 16)   # replaces W1 (4,16) + b1 (1,16)
        self.layer2 = nn.Linear(16, 3)   # replaces W2 (16,3) + b2 (1,3)
        self.relu   = nn.ReLU()           # replaces your relu() function

    def forward(self, x):
        # Exactly mirrors your NumPy forward pass
        z1 = self.layer1(x)    # x @ W1 + b1
        a1 = self.relu(z1)     # relu(z1)
        z2 = self.layer2(a1)   # a1 @ W2 + b2
        # NOTE: no softmax here — CrossEntropyLoss applies it internally
        return z2

model = IrisNet()
print("\nModel architecture:")
print(model)

# Count parameters — compare to NumPy where you tracked shapes manually
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params}")
# Expect: 4*16 + 16 + 16*3 + 3 = 64 + 16 + 48 + 3 = 131

# ─── 3. LOSS & OPTIMIZER ──────────────────────────────────────────────────────
# In NumPy: you wrote cross_entropy_loss() and gradient descent manually
# In PyTorch: one line each

criterion = nn.CrossEntropyLoss()        # replaces your cross_entropy_loss()
optimizer = optim.SGD(model.parameters(), lr=0.1)  # replaces W -= lr * dW

# ─── 4. TRAINING LOOP ─────────────────────────────────────────────────────────
# Compare to NumPy: you had forward() → loss → backward() → update weights
# PyTorch: same 4 steps, but autograd handles all the gradient math

epochs       = 1000
loss_history = []

for epoch in range(epochs):
    model.train()

    # Step 1: forward pass
    y_pred = model(X_train)

    # Step 2: compute loss
    loss = criterion(y_pred, y_train)
    loss_history.append(loss.item())

    # Step 3: backward pass — autograd computes ALL gradients automatically
    # replaces your entire backward() function (dW1, db1, dW2, db2)
    optimizer.zero_grad()   # clear old gradients first
    loss.backward()         # compute new gradients

    # Step 4: update weights
    optimizer.step()        # W -= lr * W.grad for every parameter

    if epoch % 100 == 0:
        print(f"Epoch {epoch:4d} | Loss: {loss.item():.4f}")

# ─── 5. EVALUATE ACCURACY ─────────────────────────────────────────────────────

model.eval()
with torch.no_grad():   # disable gradient tracking — not needed for inference
    train_preds = torch.argmax(model(X_train), dim=1)
    test_preds  = torch.argmax(model(X_test),  dim=1)

train_acc = (train_preds == y_train).float().mean().item()
test_acc  = (test_preds  == y_test).float().mean().item()

print(f"\nTrain accuracy: {train_acc * 100:.1f}%")
print(f"Test  accuracy: {test_acc  * 100:.1f}%")

# ─── 6. PLOT LOSS CURVE ───────────────────────────────────────────────────────

plt.figure(figsize=(8, 4))
plt.plot(loss_history, color='orange', label='PyTorch')
plt.title("Training Loss over Epochs — PyTorch")
plt.xlabel("Epoch")
plt.ylabel("Cross-Entropy Loss")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
